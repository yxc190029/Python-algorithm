```python
#學習
class Solution:
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        h = {}
        for i, num in enumerate(nums):
            n = target - num
            if n not in h:
                h[num] = i
            else:
                return [h[n], i]
#在[3,3]裡,因為h永遠放不滿,所以"目前的",跟"字典的"會永遠為有效答案,（不為同一位置同一數）,
#我的方法是先放滿,再檢查,在除bug(差diff,跟目前迴圈的num一樣,回報錯的num)...他的較省時.
```

```python
#leetcode21
#A node is a basic unit of a data structure, such as a linked list or tree data structure.
# Input: 1->2->4, 1->3->4
# Output: 1->1->2->3->4->4
k = input().split(",")
#根據輸入的資料分割,分割後同屬一個list
w = []
#str process
for i in k:
    j = i.replace("->",",").split(",")
    #字串也可以for loop,但是 容易算入符號跟空格
    for x in range(len(j)):
        w.append(j[x].strip())
w = sorted(w)
print("->".join(w))
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None
#           
#        物件導向概念
class Solution:
    def mergeTwoLists(self, l1, l2) :
        curr = dummy = ListNode(0)
        curr -> val = 1, next = 1 => 0 -> 1
        dummy -> val = 0, next None
        l1(1) -> val = 1, next = 2
        l2(1) -> val = 1, next = 3
        #初始化
        #dummy 虛擬位置,永遠在頭
        while l1 and l2:
            #L1 AND L2 exist, run
            if l1.val < l2.val:
                curr.next = l1
                l1 = l1.next
                #指向下一個
            else:
                curr.next = l2
                l2 = l2.next
            curr = curr.next

        curr.next = l1 or l2
        return dummy.next
```
