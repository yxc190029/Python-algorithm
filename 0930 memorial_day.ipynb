```python
#虔誠的基督徒
from datetime import datetime
from datetime import timedelta
k = 1
C = int(input())

for i in range(1,7):
    A = datetime(C,1,k)
    B = A.strftime("%w")
    if B == "0":
        E = A.strftime("%Y-%m-%d")
        print(E)
        for j in range (52):
            
            A += timedelta(days=7)
            E = A.strftime("%Y-%m-%d")
            G = A.timestamp()
            F = datetime(C+1,1,1).timestamp()
            if G<F:
                print(E)

    else:
        k += 1
```
